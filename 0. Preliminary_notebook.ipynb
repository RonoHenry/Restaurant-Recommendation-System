# YELP DATASET
***
The initial data was gotten from [YELP](https://www.yelp.com/dataset/download), from which we based our project on.
The dataset contained 5 JSON files from which 2 were picked to use on our academic recommendation systems project. The datasets chosen were on:-  
-  Business Information
-  User Information

```python
# importing libraries
import pandas as pd
import tarfile
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Specify the path to your tar file
tar_path = 'yelp_dataset.tar'

# Open the tar file and save to a hidden file
with tarfile.open(tar_path, 'r') as tar:
    # Extract all contents to a specific directory
    tar.extractall(path='.extracted_data')
```

***
## FIRST DATASET CONTAINING BUSINESS INFORMATION
***

```python
# File Path to JSON file
file_path = '.extracted_data/yelp_academic_dataset_business.json'

# Load the JSON file into a DataFrame,
business = pd.read_json(file_path, lines=True)

# Displaying first 5 entries
business.head()
```

```python
business.info()
```

```python
# Filter the DataFrame to keep only rows that contain the word "restaurant"
restaurants_df = business[business['categories'].str.contains("restaurant", case=False, na=False)]

# Display the first few rows of the filtered DataFrame
restaurants_df.head()
```

```python
restaurants_df.info()
```

```python
# Saving restaurant data to csv 
restaurants_df.to_csv('data/restaurants.csv', index=False)
```

***
## SECOND DATASET CONTAINING REVIEWS INFORMATION
***

```python
# File Path to  JSON file
file_path = '.extracted_data/yelp_academic_dataset_review.json'

# Define a chunk size
chunk_size = 10000

# Initialize an empty list to hold chunks
chunks = []

# Read the JSON file in chunks
with pd.read_json(file_path, lines=True, chunksize=chunk_size) as reader:
    for chunk in reader:
        # Process each chunk (e.g., filter, aggregate)
        chunks.append(chunk)

# Concatenate all chunks into a single DataFrame
reviews = pd.concat(chunks, ignore_index=True)
```

```python
# information on the reviews dataset
reviews.info()
```

```python
# Getting the unique values in the restaurant_df's business Id column
restaurants_ids = restaurants_df['business_id'].unique()
len(restaurants_ids)
```

```python
# Filtering the reveiws dataframe to include only reviews related to restaurants
rest_reviews = reviews[reviews['business_id'].isin(restaurants_ids)]
```

```python
# Getting resturant reviews infor
rest_reviews.info()
```

```python
 # first 5 entries in restaurant reviews
rest_reviews.head()
```

```python
# TImeframe of the dataset
rest_reviews.date.min(), rest_reviews.date.max()
```

```python
# Starting date of the dataframe to get 1 year worth of data
start_date = '2021-01-01 00:00:00'

# Filter the DataFrame
one_year_reviews = rest_reviews[(rest_reviews['date'] >= start_date)]

# first five entries
one_year_reviews.head()
```

```python
# Getting 1 years worth of reviews
one_year_reviews.info()
```

```python
# saving the one_year_reviews to users.csv
one_year_reviews.to_csv('data/users.csv', index= False)
```
